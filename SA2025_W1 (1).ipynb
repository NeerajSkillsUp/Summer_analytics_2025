### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg=df['mpg'].median()
above_median_mpg=df[df['mpg']>median_mpg]
above_median_mpg['hp_per_wt']=above_median_mpg['horsepower']/above_median_mpg['weight']
best_ratio=above_median_mpg['hp_per_wt'].max()
best_cars=above_median_mpg[above_median_mpg['hp_per_wt']==best_ratio]
best_cars
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg=df.groupby(['model_year','origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(10,6))
for origin in avg_mpg['origin'].unique():
    subset=avg_mpg[avg_mpg['origin']==origin]
    plt.plot(subset['model_year'],subset['mpg'],
             marker='o',
             label=origin)
plt.xlabel('model_Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Origin')
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
df['model']=df['name'].str.split().str[0]
model_years=df.groupby('model')['model_year'].nunique().reset_index()
multi_year_models=model_years[model_years['model_year']>1]['model']
filtered=df[df['model'].isin(multi_year_models)]
consistency=filtered.groupby('model').agg(
    appearances=('mpg','size'),
    std_mpg=('mpg',np.std),
    avg_mpg=('mpg','mean')
).reset_index()
consistent_cars = consistency[(consistency['std_mpg']<1.0) & (consistency['appearances']>=2)]
result=consistent_cars.sort_values(by=['appearances','avg_mpg'],ascending=[False,False])
result
```

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = sns.load_dataset('mpg').dropna()
df['model']=df['name'].str.split(n=2).str[0]+' '+df['name'].str.split(n=2).str[1]
df['model']=df['model'].str.strip()
model_stats=df.groupby('model').agg(
    years=('model_year','nunique'),
    records=('model_year','count'),
    std_mpg=('mpg',np.std),
    avg_mpg=('mpg','mean')
).reset_index()
consistent_models=model_stats[
    (model_stats['years']>1) &
    (model_stats['std_mpg']<1.0) &
    (model_stats['records']>=2)
]
result=consistent_models.sort_values(
    by=['years','avg_mpg'],
    ascending=[False,False]
)
result[['model','years','avg_mpg','std_mpg']].rename(
    columns={'years':'appearances'}
)
```
